```python
x_start = []
y_start = []
type_def = []
minute = []
half = []
team = []
game_num = []
```

```python
for x in matches:
    tree = et.ElementTree(file = x)
    gameFile = tree.getroot()
    team_dict = {gameFile[0].attrib["home_team_id"]: gameFile[0].attrib["home_team_name"],
                 gameFile[0].attrib["away_team_id"]: gameFile[0].attrib["away_team_name"]}

    for game in gameFile:
    
        for event in game:
        
            if event.attrib.get("type_id") == '7':
            
                x_start.append(event.attrib.get("x"))
                y_start.append(event.attrib.get("y"))
                type_def.append('Tackle')
                minute.append(event.attrib.get("min"))
                half.append(event.attrib.get("period_id"))
                team.append(team_dict[event.attrib.get("team_id")])
                game_num.append(x)
```

```python
for x in matches:
    
    tree = et.ElementTree(file = x)
    gameFile = tree.getroot()
    team_dict = {gameFile[0].attrib["home_team_id"]: gameFile[0].attrib["home_team_name"],
                 gameFile[0].attrib["away_team_id"]: gameFile[0].attrib["away_team_name"]}

    for game in gameFile:

        for event in game:       

            if event.attrib.get("type_id") == '8':
            
                x_start.append(event.attrib.get("x"))
                y_start.append(event.attrib.get("y"))
                type_def.append('Interception')
                minute.append(event.attrib.get("min"))
                half.append(event.attrib.get("period_id"))
                team.append(team_dict[event.attrib.get("team_id")])
                game_num.append(x)
```

```python
column_titles = ["team", "half", "min", "x_start", "y_start", "type_def", "game_num"]
final_table = pd.DataFrame(data=[team, half, minute, x_start, y_start, type_def, game_num],
                           index=column_titles)
final_table = final_table.T
final_table
```
